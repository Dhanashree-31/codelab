# one hot encoding



```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df
```

```python
dfle = df
dfle.town =  le.fit_transform(dfle.town)
dfle
```

```python
x = dfle[['town','area']].values
x
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#ohe = OneHotEncoder(categories='auto')
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [0])],   # The column numbers to be transformed (here is [0] but can be [0, 1, 3])
    remainder='passthrough'                                         # Leave the rest of the columns untouched
)
```

```python
X = ct.fit_transform(x)
X
```

```python
#to avoid dummy variable trap
#drop one column
X= X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.predict([[1,0,2800]])
```

```python
model.score(X,y)
```

# OneHotEncoding (exercise)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df
```

```python
dfle = df
dfle.car_model = le.fit_transform(dfle.car_model)
dfle
```

```python
X = dfle[['car_model','mileage','age']].values 
```

```python
X
```

```python
y = dfle.sell_price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('car_model',OneHotEncoder(), [0])],remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,2:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,45000,4]])
```

```python
model.predict([[0,86000,7]])
```

```python
model.score(X,y)
```
